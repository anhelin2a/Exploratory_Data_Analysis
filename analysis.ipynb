### Wprowadzenie

Celem projektu była analiza zależności między stopą bezrobocia a odsetkiem samobójstw w poszczególnych województwach Polski na przestrzeni lat. Badanie oparto na danych statystycznych dotyczących bezrobocia rejestrowanego (uwzględniających podział na płeć) oraz danych o zgonach z powodu samobójstw, pochodzących z GUS. Analiza objęła okres od 1999 do 2024 roku, co pozwoliło uchwycić długoterminowe trendy oraz wpływ kluczowych zdarzeń, takich jak pandemia COVID-19. Projekt miał na celu zweryfikowanie hipotezy o istotnej korelacji między bezrobociem a zdrowiem psychicznym społeczeństwa, a także wskazanie czynników łagodzących lub nasilających te zjawiska. Wykresy map zostały wykonane w narzędziu Tableau

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

Przeanalizujmy jak na przestrzeni lat zmieniał się udział bezrobotnych absolwentów wśród bezrobotnych ogółem i zbadajmy potencjalne różnice dla płci

```python
df_pivot = pd.read_excel('Udział_zarejestrowanych_bezrobotnych_absolwentów_wśród_bezrobotnych_ogółem_wg.xlsx')

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_pivot, x='Year', y='Total', marker='o', color='blue', label='Total')
plt.title('Trendy w czasie dla absolwentów ogólem')
plt.xlabel('Rok')
plt.ylabel('Stopa bezrobocia (%)')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_pivot, x='Year', y='Male', marker='o', color='green', label='Mężczyźni')
sns.lineplot(data=df_pivot, x='Year', y='Female', marker='o', color='red', label='Kobiety')
plt.title('Porównanie bezrobocia między mężczyznami a kobietami')
plt.xlabel('Rok')
plt.ylabel('Stopa bezrobocia (%)')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_pivot, x='Year', y='Male', marker='o', color='green', label='Mężczyźni')
sns.lineplot(data=df_pivot, x='Year', y='Female', marker='o', color='red', label='Kobiety')
sns.lineplot(data=df_pivot, x='Year', y='Total', marker='o', color='blue', label='Ogółem')
plt.title('Trendy w czasie dla każdej płci osobno')
plt.xlabel('Rok')
plt.ylabel('Stopa bezrobocia (%)')
plt.legend()
plt.show()
```

Ogólny trend udziału bezrobotnych absolwentów:
- Udział absolwentów w bezrobociu ogółem wykazuje spadkowy trend na przestrzeni lat.
- W 2012 roku procent bezrobotnych absolwentów wynosiło około 5.9%, a do 2021 roku spadło do około 3.1%.
- Największy spadek nastąpił między 2013 a 2014 rokiem, gdzie bezrobocie zmniejszyło się o około 1.1%.
- W latach 2021–2022 widoczny jest wzrost liczby bezrobotnych absolwentów, co może być związane z trudnościami w społeczeństwie po pandemii.


Różnice między płciami
- W większości wśród bezrobotnych absolwentów kobiety miały wyższą stopę bezrobocia niż mężczyźni.
- Największa różnica między płciami wystąpiła w 2012 roku, gdzie bezrobocie kobiet było o 1.1 punktu procentowego wyższe niż mężczyzn.
- W latach 2020–2021 nastąpiła istotna zmiana: bezrobocie kobiet spadło poniżej poziomu bezrobocia mężczyzn. W 2022 roku bezrobocie kobiet wyniosło 4.2%, podczas gdy u mężczyzn 4.7%. Ta zmiana może wynikać z większej elastyczności kobiet na rynku pracy.


Najważniejsze zmiany
- 2013–2014: W tym okresie nastąpił znaczący spadek udziłu absolwentów w bezrobociu zarówno wśród mężczyzn, jak i kobiet.
- 2020–2021: Czasy pandemii COVID-19, spodziewalibyśmy wzrostu liczby bezrobotnych osób, natomiast obserwujemy podalszy spadek
- 2022: Po pandemii bezrobocie wśród kobiet spadło poniżej poziomu bezrobocia mężczyzn, w obu przypadkach obserwujemy znaczny wzrost

Wnioski:

Najniższa stopa bezrobocia została odnotowana w 2019 roku, co może świadczyć o skuteczności działań mających na celu ograniczenie bezrobocia w tym okresie.

2. Rozkład stopy napływu bezrobotnych zarejestrowanych na przestrzeni lat
Wykres przedstawia stopę napływu bezrobotnych zarejestrowanych w procentach na przestrzeni lat od 2013 do 2024. Stopa napływu bezrobotnych odnosi się do liczby nowo zarejestrowanych bezrobotnych w stosunku do ogólnej liczby osób aktywnych zawodowo.

Trendy w czasie:
- W latach 2013–2016 stopa napływu bezrobotnych była stosunkowo wysoka, oscylując wokół 1–1.5%. Ten wysoki poziom mógł być związany z trudną sytuacją na rynku pracy w tym okresie.
- W latach 2013–2021 nastąpił stopniowy spadek stopy napływu bezrobotnych do poziomu około 0.75%. Ten spadek może świadczyć o poprawie sytuacji gospodarczej oraz wzroście zatrudnienia.
- W 2020 roku stopa napływu bezrobotnych nieznacznie wzrosła, co prawdopodobnie było skutkiem pandemii COVID-19, która wpłynęła na rynek pracy, prowadząc do zwolnień i trudności w znalezieniu nowej pracy.
- W latach 2021–2024 stopa napływu bezrobotnych utrzymuje się na niskim poziomie, oscylując wokół 1%. Może to świadczyć o stabilizacji na rynku pracy oraz skuteczności programów wsparcia dla osób bezrobotnych.

Wnioski:
Ogólny trend stopy napływu bezrobotnych jest spadkowy, co może wskazywać na poprawę sytuacji na rynku pracy oraz skuteczność działań mających na celu ograniczenie bezrobocia.

Pandemia COVID-19 miała krótkotrwały negatywny wpływ na stopę napływu bezrobotnych, ale w kolejnych latach sytuacja wróciła do stabilizacji.

3. Heatmapa stopy bezrobocia rejestrowanego na przestrzeni lat i miesięcy
Heatmapa przedstawia stopę bezrobocia rejestrowanego w procentach w poszczególnych miesiącach i latach od 2011 do 2024. Wykres ten pozwala na analizę sezonowych wahań stopy bezrobocia.

Trendy w czasie:
- W latach 2011–2012 stopa bezrobocia była najwyższa, osiągając wartości w zakresie 12,8-14,6%. Najwyższe wartości odnotowano w styczniu i lutym (około 14.5%), co może być związane z sezonowym spadkiem aktywności gospodarczej po okresie świątecznym.
- W latach 2014–2016 stopa bezrobocia spadła do wartości w zakresie 6–8%. Najniższe wartości odnotowano w czerwcu i lipcu (około 5–6%), co może wynikać z sezonowego wzrostu zatrudnienia w sektorach takich jak turystyka i rolnictwo.
- W latach 2017–2020 stopa bezrobocia utrzymywała się na niskim poziomie, oscylując wokół 3–5%. Najniższe wartości odnotowano w 2019 roku.
- W latach 2021–2024 stopa bezrobocia pozostaje na niskim poziomie, ale widoczne są niewielkie wahania między miesiącami. Najwyższe wartości nadal występują w styczniu i lutym, co sugeruje, że sezonowość nadal ma wpływ na rynek pracy.

### Podsumowanie

Analiza danych wskazuje na istotny związek między stopą bezrobocia a odsetkiem samobójstw w poszczególnych regionach Polski. Województwa o wyższym bezrobociu, takie jak warmińsko-mazurskie (8,088%) czy zachodniopomorskie (6,618%), odnotowały również wyższe wskaźniki samobójstw (odpowiednio 1,6344% i 1,7333%). Z kolei regiony z niższym bezrobociem, takie jak śląskie (4,026%) czy wielkopolskie (3,273%), charakteryzują się niższymi odsetkami samobójstw (1,0331% i 1,4930%).

Jednak nie jest to zależność jednoznaczna. Przykładem jest województwo lubuskie, gdzie mimo umiarkowanego bezrobocia (5,420%) odnotowano najwyższy wskaźnik samobójstw (1,8034%). Sugeruje to, że oprócz bezrobocia, kluczowe znaczenie mają inne czynniki, takie jak dostęp do opieki zdrowotnej, izolacja społeczna czy migracje młodych osób.

Podsumowując, bezrobocie jest istotnym, ale nie jedynym czynnikiem wpływającym na statystyki samobójstw. W regionach o trudnej sytuacji ekonomicznej ryzyko jest wyższe, ale lokalne uwarunkowania społeczne i infrastrukturalne również odgrywają ważną rolę.
